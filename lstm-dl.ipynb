```python
# Importing necessary libraries
!pip install GPUtil
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import psutil
import GPUtil
import platform
import tensorflow as tf
import time
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, roc_auc_score, roc_curve, precision_score, recall_score, f1_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import AdamW
from tensorflow.keras.regularizers import l2
from tensorflow.keras.callbacks import Callback, EarlyStopping, ReduceLROnPlateau, ModelCheckpoint

# Print environment information and system details
def print_environment_info():
    print("\nSystem and Package Versions:")
    print(f"Python Version: {platform.python_version()}")
    print(f"TensorFlow Version: {tf.__version__}")
    print(f"Pandas Version: {pd.__version__}")
    print(f"Numpy Version: {np.__version__}")
    print(f"Operating System: {platform.system()} {platform.release()}")

    # CPU information
    print("\nCPU Information:")
    print(f"Physical Cores: {psutil.cpu_count(logical=False)}")
    print(f"Total Cores: {psutil.cpu_count(logical=True)}")
    print(f"Max Frequency: {psutil.cpu_freq().max:.2f}Mhz")
    print(f"Min Frequency: {psutil.cpu_freq().min:.2f}Mhz")
    print(f"Current Frequency: {psutil.cpu_freq().current:.2f}Mhz")
    print(f"Total CPU Usage: {psutil.cpu_percent(interval=1)}%")

    # RAM information
    memory = psutil.virtual_memory()
    print("\nMemory Information (RAM):")
    print(f"Total: {get_size(memory.total)}")
    print(f"Available: {get_size(memory.available)}")
    print(f"Used: {get_size(memory.used)}")
    print(f"Percentage: {memory.percent}%")

    # Disk usage
    disk = psutil.disk_usage('/')
    print("\nDisk Information:")
    print(f"Total Size: {get_size(disk.total)}")
    print(f"Used: {get_size(disk.used)}")
    print(f"Free: {get_size(disk.free)}")
    print(f"Percentage: {disk.percent}%")

    # GPU information
    gpus = GPUtil.getGPUs()
    if len(gpus) > 0:
        print("\nGPU Information:")
        for gpu in gpus:
            print(f"GPU: {gpu.name}")
            print(f"Total Memory: {gpu.memoryTotal}MB")
            print(f"Available Memory: {gpu.memoryFree}MB")
            print(f"Used Memory: {gpu.memoryUsed}MB")
            print(f"GPU Load: {gpu.load * 100}%")
            print(f"Temperature: {gpu.temperature} °C")
    else:
        print("No GPU found.")

# Helper function to convert bytes to GB/MB
def get_size(bytes, suffix="B"):
    factor = 1024
    for unit in ["", "K", "M", "G", "T", "P"]:
        if bytes < factor:
            return f"{bytes:.2f}{unit}{suffix}"
        bytes /= factor

# Callback for better visualization and time tracking
class TrainingProgressCallback(Callback):
    def on_train_begin(self, logs=None):
        self.train_start_time = time.time()
        print("Starting training...\n")

    def on_epoch_begin(self, epoch, logs=None):
        self.epoch_start_time = time.time()
        
    def on_epoch_end(self, epoch, logs=None):
        epoch_duration = time.time() - self.epoch_start_time
        print(f"Epoch {epoch + 1}/{self.params['epochs']} - "
              f"{int(epoch_duration * 1000)}ms/epoch - loss: {logs.get('loss'):.4f} - accuracy: {logs.get('accuracy'):.4f} - "
              f"val_loss: {logs.get('val_loss'):.4f} - val_accuracy: {logs.get('val_accuracy'):.4f}\n")
        
    def on_train_end(self, logs=None):
        self.total_train_time = time.time() - self.train_start_time
        print(f"Total Training Time: {self.total_train_time:.2f} seconds\n")

# Data Preprocessing
def load_and_preprocess_data(file_path):
    data = pd.read_csv(file_path)
    
    # Drop irrelevant columns
    columns_to_drop = ['target_dos', 'target_slowite', 'target_malformed', 'target_legitimate', 'target_flood', 'target_bruteforce']
    data_filtered = data.drop(columns=columns_to_drop)

    # Split into features and target
    X = data_filtered.drop(columns=['target']).values
    y = data_filtered['target'].values  # Keep label-encoded targets
    
    # Normalize features
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    
    return X_scaled, y

# Model Architecture
def create_custom_lstm_model(input_shape):
    model = Sequential([
        LSTM(50, activation='relu', input_shape=input_shape, return_sequences=True),  # Input LSTM Layer
        LSTM(50, activation='relu', return_sequences=False),  # First Hidden LSTM Layer
        Dropout(0.25),  # Dropout layer
        Dense(40, activation='relu'),  # Dense layer with 40 units
        Dense(25, activation='relu'),  # Dense layer with 25 units
        Dense(6, activation='softmax')  # Output layer with 7 units (softmax activation for multiclass classification)
    ])
    
    # Using AdamW instead of Adam
    optimizer = AdamW(learning_rate=0.001)
    model.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

# Model Training and Evaluation
def train_and_evaluate_model(model_fn, X, y, batch_size, epochs, model_params):
    kfold = KFold(n_splits=5, shuffle=True, random_state=42)
    results = []
    fold_no = 1
    
    for train_index, val_index in kfold.split(X):
        X_train, X_val = X[train_index], X[val_index]
        y_train, y_val = y[train_index], y[val_index]
        
        input_shape = (X_train.shape[1], 1)
        X_train_reshaped = X_train.reshape(-1, X_train.shape[1], 1)
        X_val_reshaped = X_val.reshape(-1, X_val.shape[1], 1)
        
        model = model_fn(input_shape=input_shape)
        
        # Early stopping with static patience
        early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
        reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=0.0001)
        
        # Save the best model
        model_checkpoint = ModelCheckpoint(f"best_model_fold_{fold_no}.keras", monitor='val_loss', save_best_only=True, mode='min')

        # Training the model with custom callback
        progress_callback = TrainingProgressCallback()
        start_train_time = time.time()
        history = model.fit(
            X_train_reshaped, y_train,
            validation_data=(X_val_reshaped, y_val),
            epochs=epochs,
            batch_size=batch_size,
            callbacks=[early_stopping, reduce_lr, model_checkpoint, progress_callback],
            verbose=0
        )
        training_time = time.time() - start_train_time
        
        # Measure testing time
        start_test_time = time.time()
        test_loss, test_accuracy = model.evaluate(X_val_reshaped, y_val)
        testing_time = time.time() - start_test_time
        
        # Predict and calculate metrics
        y_pred_probs = model.predict(X_val_reshaped)
        y_pred_classes = np.argmax(y_pred_probs, axis=1)
        
        precision = precision_score(y_val, y_pred_classes, average='weighted')
        recall = recall_score(y_val, y_pred_classes, average='weighted')
        f1 = f1_score(y_val, y_pred_classes, average='weighted')
        roc_auc = roc_auc_score(y_val, y_pred_probs, multi_class='ovr')
        
        # Print system information
        print_environment_info()
        
        # Save metrics and losses for comparison
        metrics = {
            'fold': fold_no,
            'batch_size': batch_size,
            'epochs': epochs,
            'model_params': model_params,
            'accuracy': test_accuracy,
            'precision': precision,
            'recall': recall,
            'f1': f1,
            'roc_auc': roc_auc,
            'training_loss': history.history['loss'][-1],
            'validation_loss': history.history['val_loss'][-1],
            'training_time': training_time,
            'testing_time': testing_time
        }
        
        print(f"Fold {fold_no} - Training Time: {training_time:.2f} seconds")
        print(f"Fold {fold_no} - Testing Time: {testing_time:.2f} seconds\n")
        
        # Plot accuracy, loss, confusion matrix, and ROC curve
        plot_and_save_results(metrics, f"Fold_{fold_no}", history, y_val, y_pred_probs, y_pred_classes)
        
        results.append(metrics)
        fold_no += 1
    
    return results

# Plotting and Saving Results
def plot_and_save_results(metrics, model_name, history, y_val, y_pred_probs, y_pred_classes):
    l2_strength = 1e-4
    width_factor = 1
    
    # Plot accuracy and loss over epochs
    if history:
        plt.figure(figsize=(12, 4))
        
        # Accuracy plot
        plt.subplot(1, 2, 1)
        plt.plot(history.history['accuracy'], label='Train Accuracy')
        plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
        plt.title(f'{model_name} - Accuracy (L2: {l2_strength}, Width Factor: {width_factor})')
        plt.xlabel('Epochs')
        plt.ylabel('Accuracy')
        plt.legend()
        
        # Loss plot
        plt.subplot(1, 2, 2)
        plt.plot(history.history['loss'], label='Train Loss')
        plt.plot(history.history['val_loss'], label='Validation Loss')
        plt.title(f'{model_name} - Loss (L2: {l2_strength}, Width Factor: {width_factor})')
        plt.xlabel('Epochs')
        plt.ylabel('Loss')
        plt.legend()
        
        plt.show()
    
    # Confusion Matrix and ROC Curve
    if y_val is not None and y_pred_probs is not None and y_pred_classes is not None:
        # Confusion Matrix
        cm = confusion_matrix(y_val, y_pred_classes)
        plt.figure(figsize=(8, 6))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
        plt.title(f'{model_name} - Confusion Matrix (L2: {l2_strength}, Width Factor: {width_factor})')
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.show()
        
        # ROC AUC Score and Curves
        roc_auc = roc_auc_score(y_val, y_pred_probs, multi_class='ovr')
        fpr = {}
        tpr = {}
        
        plt.figure(figsize=(8, 6))
        for i in range(y_pred_probs.shape[1]):
            fpr[i], tpr[i], _ = roc_curve(y_val, y_pred_probs[:, i], pos_label=i)
            plt.plot(fpr[i], tpr[i], label=f'Class {i} (AUC = {roc_auc:.2f})')
        
        plt.plot([0, 1], [0, 1], 'k--')
        plt.title(f'{model_name} - ROC Curve (L2: {l2_strength}, Width Factor: {width_factor})')
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.legend(loc='best')
        plt.show()

# Main Script Execution
if __name__ == "__main__":
    file_path = '/kaggle/input/dataset/data_MQTT_final.csv'
    X, y = load_and_preprocess_data(file_path)
    
    # Model parameters
    model_params = {
        'width_factor': 1,
        'dropout_rate': 0.3,
        'learning_rate': 0.001,
        'l2_strength': 1e-4
    }
    
    # Train and Evaluate LSTM Model with the updated architecture and AdamW optimizer
    custom_training_params = {'batch_size': 300, 'epochs': 150}
    results = train_and_evaluate_model(create_custom_lstm_model, X, y, **custom_training_params, model_params=model_params)
    
    # Save results to CSV
    results_df = pd.DataFrame(results)
    results_df.to_csv('custom_model_metrics_comparison.csv', index=False)
    print("Model metrics comparison saved to 'custom_model_metrics_comparison.csv'.")
    
    # Display the results as a table for comparison
    comparison_df = pd.DataFrame(results)
    print("\nComparison Table:")
    print(comparison_df[['fold', 'batch_size', 'epochs', 'model_params', 'accuracy', 'precision', 'recall', 'f1', 'roc_auc', 'training_loss', 'validation_loss', 'training_time', 'testing_time']])
```
